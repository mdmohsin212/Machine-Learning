```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../dataset/WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

```python
df.info()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dropna(how='any', inplace=True)
x = df.drop(columns=['customerID', 'Churn'])
y = df['Churn']
```

```python
x = pd.get_dummies(x, columns=[
 'gender',
 'Partner',
 'Dependents',
 'PhoneService',
 'MultipleLines',
 'InternetService',
 'OnlineSecurity',
 'OnlineBackup',
 'DeviceProtection',
 'TechSupport',
 'StreamingTV',
 'StreamingMovies',
 'Contract',
 'PaperlessBilling',
 'PaymentMethod',
 ], drop_first=True, dtype=int)
x.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

std_enc = StandardScaler()
x_train = std_enc.fit_transform(X_train)
x_test = std_enc.transform(X_test)
```

### Ensemble Learning

### Adaboost

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report

model_eda = AdaBoostClassifier(n_estimators=200)
model_eda.fit(x_train, y_train)

y_pred_ada = model_eda.predict(x_test)
print(classification_report(y_test, y_pred_ada))
```

### Gradient boost

```python
from sklearn.ensemble import GradientBoostingClassifier

model_grd = GradientBoostingClassifier(n_estimators=200)
model_grd.fit(x_train, y_train)

y_pred_grd = model_grd.predict(x_test)
print(classification_report(y_test, y_pred_grd))
```

### XGBOOST

```python
y[y=='No'] = 0
y[y=='Yes'] = 1
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

y_test = y_test.astype(int)
y_train = y_train.astype(int)

std_enc = StandardScaler()
x_train = std_enc.fit_transform(X_train)
x_test = std_enc.transform(X_test)
```

```python
from xgboost import XGBClassifier

model_xgb = XGBClassifier()
model_xgb.fit(x_train, y_train)

y_pred_xgb = model_xgb.predict(x_test)
print(classification_report(y_test, y_pred_xgb))
```

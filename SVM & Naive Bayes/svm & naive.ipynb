```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../dataset/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.dropna(how='any', inplace=True)
```

```python
df.shape
```

```python
df.columns.to_list()
```

```python
df['Churn'].value_counts() / len(df) * 100
```

```python
x = df.drop(columns=['customerID', 'Churn'])
y = df['Churn']
```

```python
x.head()
```

```python
x = pd.get_dummies(x, columns=[
 'gender',
 'Partner',
 'Dependents',
 'PhoneService',
 'MultipleLines',
 'InternetService',
 'OnlineSecurity',
 'OnlineBackup',
 'DeviceProtection',
 'TechSupport',
 'StreamingTV',
 'StreamingMovies',
 'Contract',
 'PaperlessBilling',
 'PaymentMethod',
], dtype=int)
```

```python
x.head(1)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
x_train_sc
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

model = KNeighborsClassifier()

model.fit(x_train_sc, y_train)
```

```python
y_pred = model.predict(x_test_sc)
print(accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

model_dt = DecisionTreeClassifier()

model_dt.fit(x_train_sc, y_train)
```

```python
y_pred_dt = model_dt.predict(x_test_sc)
print(accuracy_score(y_test, y_pred_dt))
```

```python
print(classification_report(y_test, y_pred_dt))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(n_estimators=200)

model_rf.fit(x_train_sc, y_train)
```

```python
y_pred_rf = model_rf.predict(x_test_sc)
print(accuracy_score(y_test, y_pred_rf))
```

```python
print(classification_report(y_test, y_pred_rf))
```

### Naive Bayes

```python
from sklearn.naive_bayes import BernoulliNB

model_nb = BernoulliNB()
model_nb.fit(x_train_sc, y_train)

y_pred_nb = model_dt.predict(x_test_sc)
print(classification_report(y_test, y_pred_nb))
```

### SVM

```python
from sklearn.svm import SVC

model_sv = SVC()
model_sv.fit(x_train_sc, y_train)

y_pred_sv = model_sv.predict(x_test_sc)
print(classification_report(y_test, y_pred_sv))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression()
model_lr.fit(x_train_sc, y_train)

y_pred_lr = model_lr.predict(x_test_sc)
print(classification_report(y_test, y_pred_lr))
```
